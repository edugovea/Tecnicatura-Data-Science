```python
# Título Principal

import pandas as pd
```

```python
df = pd.read_excel('/content/titanic_M1.xlsx')
```

```python
## Sección Importante

df.head(3)
```

```python
# ---CONSIGNA A: Describir valores faltantes---

df.isnull().sum()
```

```python
# --- CONSIGNAC B: Completar Pclass ---
print("--- CONSIGNAC B: Rellenando 'pclass' ---\n")

# Usamos .fillna() para rellenar los valores nulos (NA) en una columna específica.
# Asignamos el valor 2, como se pidió.
# 'inplace=True' significa que modificamos al DataFrame original directamente.

df['pclass'] = df['pclass'].fillna(2)
print("Valores NA en 'pclass' rellenados con '2'.")
print("-------------------------------------------\n")

df.isnull().sum()
```

```python
# --- CONSIGNAC: Limpiando e imputando 'sex' ---
print("--- CONSIGNAC: Limpiando e imputando 'sex' --- \n")
# Primero, vemos qué valores únicos hay, para detectar inconsistencias.
print('Valores únicos en "sex" (antes):', df['sex'].unique())

df['sex'].count()

# Reemplazamos valores inconsistentes. Por ejemplo, si 'Mujer' aparece, lo mapeamos a 'female'.
df['sex'] = df['sex'].replace('Mujer', 'female')
df['sex'] = df['sex'].replace('Hombre', 'male')

print('\n Valores únicos en "sex" (después):', df['sex'].unique())
print('Columna "sex" limpia y completada.')
print('------------------------------------------- \n')
```

```python
# --- CONSIGNAC D: Rellenar 'Age' con el promedio ---
print("--- CONSIGNAC D: Rellenando 'age' con el promedio ---\n")

print("Valores NA en 'age' antes:", df['age'].isnull().sum())

# Calculamos el promedio (la media) de la columna 'age'.
# pandas ignora automáticamente los valores NA al calcular el promedio.
mean_age = df['age'].mean()
print(f"El promedio de edad calculado es: {mean_age:.2f} años.")

# Rellenamos todos los NA en la columna 'age' con el promedio que calculamos.
df['age'] = df['age'].fillna(mean_age)
print("Valores NA en 'age' después:", df['age'].isnull().sum())
print("Valores NA en 'age' rellenados con el promedio.")
print("\n-------------------------------------------------\n")
```

```python
# --- CONSIGNAC E: Rellenar 'Embarked' ---
print("--- CONSIGNAC E: Rellenando 'embarked' ---\n")

print("Valores NA en 'embarked' antes:", df['embarked'].isnull().sum())

# Rellenamos los datos vacíos en la columna 'embarked' con el valor 'S'.
df['embarked'] = df['embarked'].fillna('S')
print("Valores NA en 'embarked' rellenados con 'S'.")
print("-------------------------------------------\n")

# --- VERIFICACIÓN FINAL: Valores Faltantes (DESPUÉS) ---
print("--- VERIFICACIÓN: Valores Faltantes (DESPUÉS de limpiar) ---")
print("Valores NA en 'embarked' después:", df['embarked'].isnull().sum())
```
