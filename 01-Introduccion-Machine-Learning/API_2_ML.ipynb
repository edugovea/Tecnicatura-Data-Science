```python
# --- Importación de Bibliotecas ---
# Importamos pandas para la manipulación de datos.
import pandas as pd

# De scikit-learn (sklearn), importamos las herramientas para nuestro modelo:
# train_test_split: Para dividir nuestros datos de entrenamiento y prueba.
from sklearn.model_selection import train_test_split

# LogisticRegression: El modelo de Machine Learning que vamos a usar.
from sklearn.linear_model import LogisticRegression

# accuracy_score: La métrica para evaluar qué tan bueno es nuestro modelo.
from sklearn.metrics import accuracy_score
```

```python
# --- Carga de Datos ---
# Cargamos los dos archivos CSV en DataFrames de pandas.
df_train = pd.read_csv("/content/train.csv")
df_test = pd.read_csv("/content/test.csv")

print("--- Datos cargados (train.csv y test.csv) ---")

# Damos un primer vistazo a los datos de entrenamiento
print("Primeras 3 filas del dataset de ENTRENAMIENTO (df_train):")
df_train.head(3)
```

```python
#Datos en primer vistazo a los datos de testeo
print("Primeras 3 filas del dataset de TESTEO (df_test):")
df_test.head(3)
```

```python
# --- CONSIGNAC A: Describir valores faltantes ---
print("\n--- CONSIGNAC A: Valores Faltantes (ANTES) ---\n")

print("Valores faltantes en 'train.csv':")
# df.isnull().sum() nos da el total de datos NA (nulos) por columna.
# Filtramos para mostrar solo las columnas que tienen datos faltantes.

print(df_train.isnull().sum()[df_train.isnull().sum() > 0])

print('\nValores faltantes en \'test.csv\':')
print(df_test.isnull().sum()[df_test.isnull().sum() > 0])
```

```python
# --- CONSIGNAC B: Completar valores faltantes ---
#Esta es la etapa de "imputación" de datos.
print("\n--- CONSIGNA B: Imputando valores faltantes---")
```

```python
#1. Variable "Age" (Edad)
#Rellenamos la edad faltante con el promedio de edad de los datos

mean_age_train = df_train["Age"].mean()
print(f'Rellenando "Age" de train con el promedio de edad: {mean_age_train:.2f}')

mean_age_test = df_test["Age"].mean()
print(f'Rellenando "Age" de test con el promedio de edad: {mean_age_test:.2f}')

df_train["Age"] = df_train["Age"].fillna(mean_age_train)
df_test["Age"] = df_test["Age"].fillna(mean_age_test)
```

```python
from numpy._core.fromnumeric import mean
#2 Variable "Fare" (Tarifa)
#La consigna pide rellenar "Fare" en "Test"
#Usamos el promedio

mean_fare_test = df_test["Fare"].mean()
print(f'Rellenando "Fare" de test con el promedio: {mean_fare_test:.2f}')

df_test["Fare"] = df_test["Fare"].fillna(mean_fare_test)
```

```python
#3. Variable "Embarked" (Puerto de embarque)
#Rellenamos los datos faltantes con la "moda" (el valor mas comun) de los datos de entrenamiento

mode_embarked_train = df_train["Embarked"].mode()[0]
print(f'Rellenando "Embarked con la moda: {mode_embarked_train}')

df_train["Embarked"] = df_train["Embarked"].fillna(mode_embarked_train)
```

```python
#4. Variable "Cabin" (Cabina)
#--- Calculo de porcentaje faltante para "Cabin" ---

#Para train.csv
total_filas_train = len(df_train) #0 df_train_shape[0]
nulos_cabin_train = df_train["Cabin"].isnull().sum()
porcentaje_train = (nulos_cabin_train / total_filas_train) *100
print(f'Porcentaje de datos faltantes en "Cabin" de train: {porcentaje_train:.2f}%')

#Para test.csv
total_filas_train = len(df_test) #0 df_test_shape[0]
nulos_cabin_test = df_test["Cabin"].isnull().sum()
porcentaje_train = (nulos_cabin_test / total_filas_train) *100
print(f'Porcentaje de datos faltantes en "Cabin" de test: {porcentaje_train:.2f}%')

```

```python
print("Eliminando la columna 'Cabin' (demasiados valores faltantes).")
df_train = df_train.drop("Cabin", axis=1)
df_test = df_test.drop("Cabin", axis=1)

```

```python
df_test.columns
```

```python
#---Preparación de Datos para el Modelo---
#Debemos convertir las variables categóricas.

print("\n--- Preparación de Datos para el Modelo (Feature Engineering) ---")

#1. Converrtir 'Sex' (texto) a numérico (0 o 1)
#Usamos .map() para reemplazar 'male' por 0 y 'female' por 1
df_train["Sex"] = df_train["Sex"].map({"male": 0, 'female': 1})
df_test["Sex"] = df_test["Sex"].map({"male": 0, 'female': 1})

```

```python
df_train.head(2)
```

```python
df_test.head(2)
```

```python
import pandas as pd

# 1 & 2. Convertir 'Embarked' (texto) a numérico

# 2 # Usamos 'One-Hot Encoding' (pd.get_dummies) que crea nuevas columnas (Embarked_S, Embarked_C, Embarked_Q).
# 3 # con valores 1 o 0, indicando el puerto.

df_train = pd.get_dummies(df_train, columns=['Embarked'])
df_test = pd.get_dummies(df_test, columns=['Embarked'])

```

```python
df_train.head(2)
```

```python
df_test.head(2)
```

```python
# 1 # 3. Seleccionar las variables (features) que usará el modelo
# 2 # Eliminamos columnas que no aportan al modelo (como 'Name' o 'Ticket').

features_to_drop = ['PassengerId', 'Name', 'Ticket']

# Eliminamos las columnas de los dataframes
df_train_processed = df_train.drop(features_to_drop, axis=1)
df_test_processed = df_test.drop(features_to_drop, axis=1) # Preparado para predicciones futuras

print("Datos convertidos a formato numérico y variables no relevantes eliminadas.")
```

```python
# --- CONSIGNAC C: Ajustar el primer modelo de regresión logística ---
print("\n--- CONSIGNAC C: Ajustando el modelo ---")

# 1. Definir nuestras variables X (features) y Y (target)
# 'X' son las columnas que usamos para predecir.
# 'y' es la columna que 'queremos' predecir ('Survived').

X = df_train_processed.drop('Survived', axis=1)
y = df_train_processed['Survived']
```

```python
# Ahora podemos aplicar train_test_split
# 1 & 2. Dividir los datos de 'train' en dos sets:
# X_train, y_train: 80% de los datos, para entrenar el modelo.
# X_test, y_test: 20% de los datos, para probar el modelo y medir el accuracy.
# ESTO ES IMPORTANTE: NO TENEMOS LOS DATOS DE SURVIVED DE TEST
# "test_size=0.2" significa 20% para prueba.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=28)

print(f"Datos de entrenamiento divididos: {len(X_train)} para entrenar, {len(X_test)} para probar.")


# Creación del Modelo
# 1 & 3. Crear (instanciar) el modelo
# max_iter=2000 es para asegurar que el modelo tenga suficientes iteraciones para "aprender".
# Es posible que salte un warning con 1000 o menos iteraciones por las diferentes escalas en las columnas
model = LogisticRegression(max_iter=2000)
```

```python
# --- Creación del modelo ---
# 1 # 3. Crear (instanciar) el modelo
# max_iter=2000 es para asegurar que el modelo tenga suficientes iteraciones para "aprender".
model = LogisticRegression(max_iter=2000)
```

```python
# --- Entrenamos el modelo ---
# 1 # 4. Entrenar el modelo
# Usamos .fit() para que el modelo aprenda la relación entre X_train e y_train.
model.fit(X_train, y_train)

print("\nModelo entrenado exitosamente.")
```

```python
# --- CONSIGNAC D: Entrenar y determinar el nivel de accuracy --
print("\n--- CONSIGNAC D: Entrenando y determinando el nivel de accuracy ---")

# 1 # 1. Hacer predicciones en el set de prueba (el 20% que separamos)
# 5 # El modelo ahora "adivina" quién sobrevivió en el set X_test.
y_pred = model.predict(X_test)
y_pred

# Muestra el array de predicciones (y_pred)
# print(y_pred) # El código original imprimía y_pred
```

```python
# 1 # 2. Calcular el Accuracy
# 2 # Comparamos las predicciones del modelo (y_pred) con las respuestas reales (y_test).
accuracy = accuracy_score(y_test, y_pred)

# 3 # 3. Mostrar el resultado
# 6 # El accuracy es el porcentaje de predicciones correctas.
print(f"\nEl nivel de ACCURACY (precisión) del modelo es: {accuracy * 100:.2f} %")
print(f"(Esto significa que el modelo acertó el {accuracy * 100:.2f}% de las veces en el set de prueba)")
```

```python
# --- PREDICCIONES FINALES (sobre test.csv) ---
print("\n--- PREDICCIONES FINALES (sobre test.csv) ---")
print("Usando el modelo entrenado para predecir el archivo test.csv...")

# 1. Preparamos el set de test.csv
# Quitamos las columnas que no se usaron para entrenar el modelo
features_to_drop = ['PassengerId', 'Name', 'Ticket']

# Creamos una copia del dataframe de prueba antes de manipularlo, si es necesario,
# o usamos df_test_processed que ya tiene algunas eliminaciones previas.
X_final_test = df_test.drop(features_to_drop, axis=1)

# 2. Alineamos las columnas de test.csv para que coincidan EXACTAMENTE con las de entrenamiento
# Usamos 'X.columns' (definido en la Consigna C) que tiene el orden de features correcto.
# Esto es crucial por si el 'get_dummies' creó columnas en orden diferente.
X_final_test = X_final_test[X.columns]

# 3. Hacer las predicciones finales
# Usamos el 'model' que ya entrenamos con X_train, y_train
final_predictions = model.predict(X_final_test)
final_predictions
```
